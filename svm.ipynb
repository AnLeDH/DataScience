```python
import pandas as pd
import numpy as np
import prepare_data
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
#Import svm model
from sklearn import svm
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# get yearly and monthly data
df_av_yearly, df_av_monthly = prepare_data.prepare_data()

# individual regression
type = "temp_max"
month = 8
year = 2050 - 1984

# prepare x and y
y = df_av_monthly[type][df_av_monthly["month"] == month]
X = pd.DataFrame(np.arange(0,len(y))).values
```

```python
y_train = df_av_monthly[type][df_av_monthly["month"] == month].values[0:35]
X_train = X[0:35,:]

prediction_data = pd.DataFrame(np.arange(0,40)).values

clf = svm.SVR(kernel='poly') # polynomial Kernel
#Train the model using the training sets
clf.fit(X_train, y_train)
#Predict the response for test dataset
y_pred = clf.predict(prediction_data)



# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'blue')
plt.plot(prediction_data, y_pred, color = 'red')
plt.title('SVM Regression')
plt.xlabel('years starting from 1984')
plt.ylabel('maximal monthly Temperature (August)')
plt.show()

print("score: ", metrics.r2_score(y, y_pred[0:len(y)]))
```
